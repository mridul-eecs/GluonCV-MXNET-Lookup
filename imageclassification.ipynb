```python
image_url= 'https://raw.githubusercontent.com/dmlc/web-data/master/gluoncv/classification/mt_baker.jpg'
image_filename= 'mt_baker.jpg'
gcv.utils.download(url= image_url, path= image_filename)
```

```python
import matplotlib.pyplot as plt
import mxnet as mx
import numpy as np
%matplotlib inline
image= mx.image.imread('mt_baker.jpg')
print(image.shape)
plt.imshow(image.asnumpy())
```

```python

image= gcv.data.transforms.presets.imagenet.transform_eval(image)
print(image.shape)
```

```python
import inspect
inspect.getmembers(gcv.model_zoo)
```

```python
network= gcv.model_zoo.resnetv1b.resnet50_v1c(pretrained= True)
```

```python
prediction= network(image)
prediction.shape
```

```python
pred= np.argmax(np.array(prediction[0]))
network.classes[pred]
```

```python
top5= np.argsort([i.asscalar() for i in prediction[0]])[::-1][:5]
[print(network.classes[i]) for i in top5]
```
